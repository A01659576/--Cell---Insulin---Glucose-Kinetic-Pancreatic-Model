Verificación de cálculos númericos.

```python
# Incializar variables
r_0 = 864
ge = 140
e_go = 1.44
si = 0.72
sigma = 43.2
alpha = 20000
p = 0.41
k = 432
d_0 = 0.06
r_1 = 0.00084
r_2 = 0.0000024
```

```python
# Primer punto de equilibrio

# R_0 + Ge / Ego
G = (r_0 + ge) /e_go
I = 0
B = 0

print(f'El primer punto de equilibrio es ({round(G,2)}, {I}, {B})')
```

```python
# Segundo punto de equilibrio
# Encontrar raices de r_2*G**2 - r_1 * G + d_0

from sympy import symbols, solve
g = symbols('g')
G1, G2 = solve(r_2*(g**2) - r_1*g + d_0, g)

print(f'G_1: {G1}')
print(f'G_2: {G2}')

B1 = ((r_0 + ge - G1*e_go)/(si*G1)) * (((alpha + G1**2) * (p+k))/(sigma * G1**2))

B2 = ((r_0 + ge - G2*e_go)/(si*G2)) * (((alpha + G2**2) * (p+k))/(sigma * G2**2))

I1 = (G1**2 * B1 * sigma) / ((alpha + G1**2)*(p+k))

I2 = (G2**2 * B2 * sigma) / ((alpha + G2**2)*(p+k))

print(f'El segundo punto de equilibrio es ({int(G1)}, {round(I1,2)}, {round(B1,2)})')
print(f'El tercer punto de equilibrio es ({int(G2)}, {round(I2,2)}, {round(B2,2)})')
```

```python
from sympy import symbols, Matrix, diff

# Define symbolic variables
G_sym, I_sym, B_sym = symbols('G I B')

# Define the system of differential equations symbolically
f1 = r_0 + ge - (e_go + si*I_sym)*G_sym
f2 = (B_sym*(G_sym**2)*(sigma)) / (alpha + G_sym**2) - (p+k)*I_sym
f3 = (-d_0 + r_1*G_sym - r_2*G_sym**2)*B_sym
display(f1)
display(f2)
display(f3)

# Create a vector of functions and variables
F = Matrix([f1, f2, f3])
V = Matrix([G_sym, I_sym, B_sym])

# Calculate the Jacobian matrix
Jacobian_sym = F.jacobian(V)

print("Symbolic Jacobian Matrix:")
display(Jacobian_sym)
```

```python
from sympy import S

# Function to calculate numerical Jacobian at a given point
def calculate_jacobian(G_val, I_val, B_val):
    subs_dict = {
        G_sym: G_val,
        I_sym: I_val,
        B_sym: B_val,
        r_0: r_0,
        ge: ge,
        e_go: e_go,
        si: si,
        alpha: alpha,
        sigma: sigma,
        p: p,
        k: k,
        d_0: d_0,
        r_1: r_1,
        r_2: r_2
    }
    return Jacobian_sym.subs(subs_dict)

# Equilibrium Point 1: (G, I, B)
G_eq1 = G
I_eq1 = I
B_eq1 = B
Jacobian_eq1 = calculate_jacobian(G_eq1, I_eq1, B_eq1)
print(f"\nJacobian at Equilibrium Point 1 ({G_eq1, 2}, {I_eq1}, {B_eq1}):")
display(Jacobian_eq1)

# Equilibrium Point 2: (G1, I1, B1)
G_eq2 = G1
I_eq2 = I1
B_eq2 = B1
Jacobian_eq2 = calculate_jacobian(G_eq2, I_eq2, B_eq2)
print(f"\nJacobian at Equilibrium Point 2 ({G_eq2}, {I_eq2}, {B_eq2}):")


Jacobian_eq2 = Jacobian_eq2.as_mutable()
Jacobian_eq2[2, 2] = S(0)

display(Jacobian_eq2)


# Equilibrium Point 3: (G2, I2, B2)
G_eq3 = G2
I_eq3 = I2
B_eq3 = B2
Jacobian_eq3 = calculate_jacobian(G_eq3, I_eq3, B_eq3)
print(f"\nJacobian at Equilibrium Point 3 ({G_eq3}, {I_eq3}, {B_eq3}):")


Jacobian_eq3 = Jacobian_eq3.as_mutable()
Jacobian_eq3[2, 2] = S(0)

display(Jacobian_eq3)


# Find eigenvalues for each Jacobian
eigenvalues_eq1 = np.linalg.eigvals(np.array(Jacobian_eq1).astype(float))
eigenvalues_eq2 = np.linalg.eigvals(np.array(Jacobian_eq2).astype(float))
eigenvalues_eq3 = np.linalg.eigvals(np.array(Jacobian_eq3).astype(float))
print(f"\nEigenvalores punto de equilibrio  1: {eigenvalues_eq1}")
print(f"Eigenvalores punto de equilibrio  2: {eigenvalues_eq2}")
print(f"Eigenvalores punto de equilibrio  3: {eigenvalues_eq3}")
```
